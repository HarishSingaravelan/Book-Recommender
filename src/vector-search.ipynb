```python
# Store the document embeddings into vector db.
from langchain_chroma import Chroma
# Its gonna take the raw text of the description and convert it into format langchain can work with.
from langchain_community.document_loaders import TextLoader 
# Converts the chunks into document embeddings.
#from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from sentence_transformers import SentenceTransformer
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv("../data/cleaned-books.csv")
books.head()
```

```python
books["tagged_description"].to_csv("../data/tagged_description.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False)
```

```python
raw_document = TextLoader("../data/tagged_description.txt", encoding="utf-8").load()
# chunksize ensure it splits on "\n"
# chunk_overlap =0 ensure there is no chunk_overlap
text_splitter = CharacterTextSplitter(chunk_size = 0, chunk_overlap = 0, separator="\n") 
documents = text_splitter.split_documents(raw_document)
```

```python
documents[0]
```

```python
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db_books = Chroma.from_documents(
    documents,
    embedding=HuggingFaceEmbeddings()
)
```

```python
query = "productivity"
docs = db_books.similarity_search(query, k = 20)
docs
```

```python
books[books["isbn13"]==int(docs[0].page_content.split()[0].strip())]["title"]
```

```python
def retrieve_semantive_recommendation(query:str, top_k: int=10) -> pd.DataFrame:
    docs = db_books.similarity_search(query, k = top_k)
    
    books_list=[]
    for i in range(0, len(docs)):
        books_list+= [int(docs[i].page_content.strip('"').split()[0])]

    
    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantive_recommendation("A book to teach children about nature")
```
